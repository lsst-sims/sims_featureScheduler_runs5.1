```python
import sqlite3
import numpy as np
import matplotlib.pylab as plt
import pandas as pd
%matplotlib inline
```

```python
con = sqlite3.connect("fid_seeing1_v5.1.0_1yrs.db")
s1 = pd.read_sql("select * from observations;", con)
con.close()

con = sqlite3.connect("fid_seeing2_v5.1.0_1yrs.db")
s2 = pd.read_sql("select * from observations;", con)
con.close()
```

```python
b2c = {"u": "purple", "g": "blue", "r":"red", "y":"green"}

fig, ax = plt.subplots()

for bn in b2c:

    g1 = np.where(s1["band"] == bn)[0]
    y = np.arange(g1.size)
    ax.plot(s1["night"][g1], y, color=b2c[bn], label=bn)

    g2 = np.where(s2["band"] == bn)[0]
    y = np.arange(g2.size)
    ax.plot(s2["night"][g2], y, '--', color=b2c[bn])

ax.set_xlabel("Night")    
ax.set_ylabel("N observations")
ax.legend()
```
