```python
from rubin_scheduler.scheduler.utils import CurrentAreaMap
from possible_footprints import *
import healpy as hp
import matplotlib.pylab as plt
%matplotlib inline
import numpy as np
```

```python
sag = CurrentAreaMap()
n_total = 1.8e6 # Around 1.8 million total visits for the main survey footprint (so, excluding DDFs, twilight near the sun, etc)
visits_per_filter, visit_sums, labels =sag.estimate_visits(n_total)

area = np.where(visit_sums > 800)[0].size * hp.nside2pixarea(32,degrees=True)

hp.mollview(visit_sums, title='Baseline, %.0f sq deg > 800 visits' % area, unit="N visits")
```
